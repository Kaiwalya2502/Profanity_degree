# Profanity Degree in Sentence

#### 1)I have taken sample facebook comment file from kaggle and used it for the purpose
#### 2)I have assumed the profanity words set as a list in the below code which can be modified for actual profanity word set.I have taken some words as example in the profanity set
#### 3) I have used the number of profanity words occuring in a sentence as a degree of profanity.
#### 4) Degree of profanity is printed only for sentences which have profanity word in it.

```python
comments_file = open('fb_data.txt','r')
profanity_list = ['love','hi','drug','shit']
line = 0
for sentence in comments_file:
    degree = 0
    
    line = line+1
    words = sentence.split(" ")
    
    for word in words:
        word = word.lower()
        if (word in profanity_list):
            print(word)
            degree+=1
    if degree > 0 :
        print('Sentence profanity degree is :',degree,'in line number :',line)
comments_file.close()
```
